```python
#load the data analysis libraris
import numpy as np
import matplotlib.pyplot as plt
```

```python
rainfall =[77,50,80,16,100]
months =['jan','feb','mar','apr','jun']
```

```python
plt.style.available
```

```python
plt.style.use("dark_background")#theme
```

```python
plt.bar(months,rainfall)
```

```python
plt.pie(
    rainfall,
    labels=months,
    autopct='%1.1f%%',
)
plt.show()
```

```python
plt.pie(
    rainfall,
    labels=months,
    autopct='%1.1f%%',
    wedgeprops={'width': 0.4, 'edgecolor': '#6666ff','linewidth':2},
    pctdistance=.8
)
plt.show()
```

```python
plt.boxplot(rainfall,
            vert=False,
            showmeans=True,
            meanline=True,
            meanprops={'color':'red'}
            )
plt.show()
```
